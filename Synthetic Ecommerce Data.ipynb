<div style="text-align: justify; font-size: 16px; color: #000000; font-family: Garamond;">
    This project analyzes a dataset of e-commerce transactions to derive key insights and metrics that can help optimize business strategies. The dataset contains detailed information about individual transactions, including data such as:

- **Transaction ID**: Unique identifier for each transaction.
- **Customer ID**: Identifier for the customer making the purchase.
- **Product ID**: Identifier for the purchased product.
- **Transaction Date**: Date of the transaction.
- **Units Sold**: Number of units sold in the transaction.
- **Discount Applied**: Discount applied on the transaction.
- **Revenue**: Total revenue generated from the transaction.
- **Clicks**: Number of clicks on ads associated with the transaction.
- **Impressions**: Number of times ads were shown to customers.
- **Conversion Rate**: Percentage of visitors who made a purchase.
- **Category**: Product category to which the item belongs.
- **Region**: Geographic region where the transaction occurred.
- **Ad CTR (Click-Through Rate)**: Percentage of ad impressions that resulted in clicks.
- **Ad CPC (Cost Per Click)**: Cost incurred for each ad click.
- **Ad Spend**: Total money spent on advertising.

**Objective**

The objective of this analysis is to explore the relationship between different marketing metrics, sales data, and customer behavior. Through various exploratory data analysis (EDA) techniques, we aim to uncover trends, correlations, and insights that could be valuable for improving sales strategies, optimizing ad campaigns, and driving better business outcomes. Through this analysis, we aim to identify areas for improvement and offer actionable recommendations for businesses seeking to maximize the effectiveness of their marketing strategies.

```python
# Plot distribution of Revenue, Ad Spend and Units Sold
df['Revenue'].hist(bins=2000, figsize=(10, 6), color='skyblue', edgecolor='orangered')
plt.title('Distribution of Revenue')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()

df['Ad_Spend'].hist(bins=2000, figsize=(10, 6), color='darkorange', edgecolor='darkorange')
plt.title('Distribution of Ad_Spend')
plt.xlabel('Ad_Spend')
plt.ylabel('Frequency')
plt.show()

df['Units_Sold'].hist(bins=2000, figsize=(10, 6), color='lightgreen', edgecolor='#f1b04c')
plt.title('Distribution of Units Sold')
plt.xlabel('Units Sold')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plots to show outliers in Revenue, Ad Spend, and Unit Sold
sns.boxplot(x=df['Revenue'], color='orangered')
plt.title('Boxplot of Revenue')
plt.show()

sns.boxplot(x=df['Ad_Spend'], color='darkorange')
plt.title('Boxplot of Ad Spend')
plt.show()

sns.boxplot(x=df['Units_Sold'], color='orange')
plt.title('Boxplot of Units Sold')
plt.show()
```

```python
# Display the total revenue per category
category_revenue = df.groupby('Category')['Revenue'].sum()
plt.figure(figsize=(8, 8))
plt.pie(category_revenue, labels=category_revenue.index, autopct='%1.1f%%', startangle=90, 
        colors=sns.color_palette("Oranges", len(category_revenue)), wedgeprops=dict(width=0.3))
plt.title("Revenue Distribution by Category", fontsize=16)
plt.show()
```

```python
# Group by Date and sum the revenue
df.groupby('Transaction_Date')['Revenue'].sum().plot(figsize=(10, 6), color='orangered')
plt.title('Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.show()

# To plot Revenue and Ad Spend over time
df_grouped = df.groupby('Transaction_Date')[['Revenue', 'Ad_Spend']].sum()
plt.figure(figsize=(10, 6))
plt.plot(df_grouped.index, df_grouped['Revenue'], label='Revenue', color='orangered', linewidth=2)
plt.plot(df_grouped.index, df_grouped['Ad_Spend'], label='Ad Spend', color='darkorange', linewidth=2)
plt.title('Revenue and Ad Spend Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Amount ($)', fontsize=12)
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(x=df['Ad_Spend'], y=df['Conversion_Rate'], color='#FF8C00')  # Darker shade of orange
plt.title('Conversion Rate vs Ad Spend')
plt.xlabel('Ad Spend')
plt.ylabel('Conversion Rate')
plt.show()
```

```python
# Filter only the columns with numeric data types (int and float)
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Generate the correlation matrix
corr_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#REVENUE BY CATEGORY
category_performance = df.groupby('Category').agg({'Revenue': 'sum', 'Units_Sold': 'sum'}).reset_index()
category_performance = category_performance.sort_values(by='Revenue', ascending=False)

# Plot the sorted bar chart with unique colors for each bar
palette = sns.color_palette("Oranges", len(category_performance))
plt.figure(figsize=(10, 6))
bars = plt.bar(category_performance['Category'], category_performance['Revenue'], color=palette)
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f"{bar.get_height():,.2f}",
        ha='center',
        va='bottom',
        fontsize=10
    )
plt.title("Revenue by Category", fontsize=16)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Revenue", fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
# Find revenue by region
region_performance = df.groupby('Region').agg({'Revenue': 'sum', 'Units_Sold': 'sum', 'Ad_Spend': 'sum'}).reset_index()
region_performance = region_performance.sort_values(by='Revenue', ascending=False)
palette = sns.color_palette("Oranges", len(region_performance))

# Create the horizontal bar chart
plt.figure(figsize=(10, 6))
bars = plt.barh(
    region_performance['Region'], 
    region_performance['Revenue'], 
    color=palette, 
    edgecolor="white", 
    height=0.7, 
    alpha=0.9,  # Slight transparency for a sleek look
    linewidth=1.2  # Adds a bold outline
)
# Annotate each bar with the Revenue values
for bar in bars:
    plt.text(
        bar.get_width() + max(region_performance['Revenue']) * 0.01,
        bar.get_y() + bar.get_height() / 2, 
        f"{bar.get_width():,.2f}",
        va='center', 
        fontsize=10, 
        color="black"
    )
plt.title("Revenue by Region", fontsize=16, weight="bold")
plt.xlabel("Revenue", fontsize=12)
plt.ylabel("Region", fontsize=12)
plt.grid(axis='x', linestyle="--", alpha=0.6)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

This indicates that higher ad spending enhances click-through rates and cost-per-click, which may reflect increased ad visibility or competitiveness in bidding.A negative correlation (-0.51) between Impressions and Conversion_Rate suggests diminishing returns on large-scale impressions, likely due to a less targeted audience. Clicks and Conversion_Rate (0.32) have a moderate positive correlation, emphasizing the role of engagement in driving conversions but also highlighting the need for other factors to improve conversion rates. <br>
    * Revenue by Category (Bar Chart): The chart provides several insights into the revenue distribution across different product categories. Electronics leads with the highest revenue of 11,326,820.25, followed closely by Books and Toys, each generating over 11 million in revenue. Clothing and Home Appliances also contribute significantly, with revenues of 11,034,899.26 and 10,559,207.22 respectively. The relatively close revenue figures for Books and Toys indicate a competitive market, whereas Home Appliances shows potential for growth compared to the other categories. <br>
    *Revenue and Ad Spend Over Time (Line Chart): The stability of ad spend across the year ($10,000/month) compared to fluctuating revenue highlights that the revenue spike in November 2024 was influenced by factors beyond ad spending, such as seasonal campaigns or product launches. Evaluating marketing campaigns and seasonal trends during November can uncover replicable strategies to sustain revenue spikes in future periods. <br>
    To optimize performance based on the insights, it is recommended to focus on refining ad spend strategies by identifying the optimal budget range that maximizes conversion rates while minimizing variability. Investing in targeted advertising campaigns can address the negative correlation between impressions and conversion rates, ensuring ads reach the most relevant audience. For product categories, enhancing marketing efforts around Electronics, Books, and Toys, while exploring growth opportunities in Home Appliances, could drive revenue further. Regionally, doubling down on Asia's strong performance with tailored strategies while maintaining competitive efforts in North America and Europe is essential. Additionally, analyzing the November revenue spike can help identify successful elements, such as seasonal demand or effective campaigns, to replicate and sustain such peaks in the future.
